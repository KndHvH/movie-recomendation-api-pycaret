```python
import pandas as pd
import numpy as np
import openai
```

```python
def enhance_sinopse(df):
    sinopses_aprimoradas = []
    df_temp = df.copy()
    for index, row in df_temp.iterrows():
        nome_do_filme = row['title_pt']
        sinopse = row['sinopse']
        
        prompt = f"Descreva o filme {nome_do_filme} de forma mais explicativa porem de forma resumida em 150 tokens com base na seguinte sinopse: {sinopse}, e gostaria que a sua resposta fosse em portugues do brasil"
        
        response = openai.Completion.create(
        model="text-davinci-002",
        prompt=prompt,
        max_tokens=150
        )

        response = response.choices[0].text.strip()
        
        print('Filme:',nome_do_filme)
        print('Sinopse Antiga:',sinopse)
        print('Sinopse Nova:',response)
        print('--------------------------')



        sinopses_aprimoradas.append(response) 
    df_temp['sinopse_aprimorada'] = sinopses_aprimoradas
    return df_temp
        
```

```python

df = pd.read_csv('api/app/database/all_movies.csv', sep = ';')
df.head()
```

```python
df1 = df.head(1)
df1
```

```python
df2 = enhance_sinopse(df1)
```
