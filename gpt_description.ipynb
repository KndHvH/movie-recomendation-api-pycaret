```python
import pandas as pd
import numpy as np
import openai
import getpass
```

```python
API_KEY = getpass.getpass("Enter your API key: ")
openai.api_key = API_KEY
```

```python
def enhance_sinopse(df):
    sinopses_aprimoradas = []
    df_temp = df.copy()
    size = len(df)
    for index, row in df_temp.iterrows():
        nome_do_filme = row['title_pt']
        sinopse = row['sinopse']
        
        prompt = f"""
        Escreva uma melhor sinopse em portugues para o filme {nome_do_filme}
        com base na sua antiga sinopse{sinopse}
        quero que a sua resposta seja apenas a nova sinopse
        Gostaria que essa nova sinopse falasse mais sobre como e o filme
        """
        
        response = openai.Completion.create(
        model="text-davinci-002",
        prompt=prompt,
        max_tokens=200
        )
 
        response = response.choices[0].text.strip()
        
        print(f'Index:{index+1}/{size}')
        print('Filme:',nome_do_filme)
        print('Sinopse Antiga:',sinopse)
        print('Sinopse Nova:',response)
        print('--------------------------')



        sinopses_aprimoradas.append(response) 
    df_temp['sinopse_aprimorada'] = sinopses_aprimoradas
    return df_temp
        
```
